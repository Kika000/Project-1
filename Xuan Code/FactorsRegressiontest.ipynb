```python
# Dependencies
import pandas as pd
from scipy.stats import linregress
from matplotlib import pyplot as plt
```

```python
analysisdata = pd.read_csv("analysisdata.csv")[["Country","2016","Disaster rate","Internet Access Rate"]]
analysisOECDdata = pd.read_csv("analysisOECDdata.csv")[["Country","2016","Disaster rate","Internet Access Rate","Average Sleeping (Hours)"]]
```

## Run the regression for factors

```python
# Disaster rate vs Happiness
xdata1 = analysisdata["Disaster rate"]
ydata1 = analysisdata["2016"]
(slope1, intercept1,r1, _, _) = linregress(xdata1,ydata1)
fitline1 = slope1 * xdata1 + intercept1
```

```python
# Plot the relationship between disaster rate and Happyiness
analysisdata.plot.scatter(x="Disaster rate", y="2016")
plt.ylabel("Happiness Score 2016")
plt.plot(xdata1, fitline1, 'r--')
r1
```

```python
# Internet Access Rate vs Happiness
xdata2 = analysisdata["Internet Access Rate"]
ydata2 = analysisdata["2016"]
(slope2, intercept2,r2, _, _) = linregress(xdata2,ydata2)
fitline2 = slope2 * xdata2 + intercept2
```

```python
# Plot the relationship between Internet Access Rate and Happyiness
analysisdata.plot.scatter(x="Internet Access Rate", y="2016")
plt.ylabel("Happiness Score 2016")
plt.plot(xdata2, fitline2, 'r--')
r2
```

```python
# Average Sleeping (Hours) vs Happiness
xdata3 = analysisOECDdata["Average Sleeping (Hours)"]
ydata3 = analysisOECDdata["2016"]
(slope3, intercept3,r3, _, _) = linregress(xdata3,ydata3)
fitline3 = slope3 * xdata3 + intercept3
```

```python
# Plot the relationship between Average Sleeping (Hours) and Happyiness
analysisOECDdata.plot.scatter(x="Average Sleeping (Hours)", y="2016")
plt.ylabel("Happiness Score 2016")
plt.plot(xdata3, fitline3, 'r--')
r3
```

```python
# Disaster rate vs Happiness
xdata1 = analysisdata["Disaster rate"]
ydata1 = analysisdata["2016"]
(slope1, intercept1,r1, _, _) = linregress(xdata1,ydata1)
fitline1 = slope1 * xdata1 + intercept1

# Internet Access Rate vs Happiness
xdata2 = analysisdata["Internet Access Rate"]
ydata2 = analysisdata["2016"]
(slope2, intercept2,r2, _, _) = linregress(xdata2,ydata2)
fitline2 = slope2 * xdata2 + intercept2

# Average Sleeping (Hours) vs Happiness
xdata3 = analysisOECDdata["Average Sleeping (Hours)"]
ydata3 = analysisOECDdata["2016"]
(slope3, intercept3,r3, _, _) = linregress(xdata3,ydata3)
fitline3 = slope3 * xdata3 + intercept3

fig = plt.figure(figsize = (12,10))
fig.subplots_adjust(hspace=0.2, wspace=0.4)
fig.suptitle("Disaster rate, Internet access rate and Average Sleeping (Hours)", fontsize = 20)
ax1 = fig.add_subplot(2,2,1)
ax1.scatter(x=analysisdata["Disaster rate"], y=analysisdata["2016"])
plt.ylabel("Happiness Score 2016")
plt.xlabel("Disaster Rate")
plt.plot(xdata1, fitline1, 'r--')
ax1.text(15,3,"r value = "+str(r1.round(3)),fontsize = 12,color = "g")

ax2 = fig.add_subplot(2,2,2)
ax2.scatter(x=analysisdata["Internet Access Rate"], y=analysisdata["2016"])
plt.ylabel("Happiness Score 2016")
plt.xlabel("Internet Access Rate")
plt.plot(xdata2, fitline2, 'r--')
ax2.text(0.6,3,f"r value = {r2.round(3)}",fontsize = 12,color = "g")

ax3 = fig.add_subplot(2,2,3)
ax3.scatter(x=analysisOECDdata["Average Sleeping (Hours)"], y=analysisOECDdata["2016"])
plt.ylabel("Happiness Score 2016")
plt.xlabel("Average Sleeping (Hours) (OECD Countries)")
plt.plot(xdata3, fitline3, 'r--')
ax3.text(7.8,4.7,f"r value = {r3.round(3)}",fontsize = 12,color = "g")

plt.savefig("../ChartsTables/regression_analysis.png")
plt.show()
```
